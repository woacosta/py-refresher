```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("darkgrid")
# darkgrid, whitegrid, dark, white, and ticks
# plt.style.use('fivethirtyeight')
# https://medium.com/@andykashyap/top-5-tricks-to-make-plots-look-better-9f6e687c1e08
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## Distplots (univariate)
set kde=false to get histogram

```python
sns.distplot(tips['total_bill'],kde=False,bins=15)
```

## Jointplot (bivariate)
basically two distplots together with a scatter plot. use the KIND argument to change to hex or regression=reg or kde. scatter is default

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

```python
# note this is how you use matplotlib to create the same scatterplot
plt.scatter(tips['total_bill'],tips['tip'])
```

## Pairplot
plots pairwise relationships for all numerical columns of dataframe. change the hue to add categorical column detail. use palette to change colour scheme

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

## Rugplot
drop a dash for every entry of the variable. can use it to build kernel density estimation (kde). basically sum of all normal distributions centred around every point of the variable.

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

## Bar plots
categorical vs numerical variables. shows mean by default

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

#### use the estimator key to use statistical functions on the categorical groups.
ex. find the standard deviation of total bill for male and female

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.max)
```

## Strip plot

## swarm plot
like a strip plot but points are adjusted to not overlap. like a violin and strip plot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

### Combine swarm and violin

```python
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
```

# Matrix plots and heatmaps
### heatmap colors https://matplotlib.org/3.1.1/gallery/color/colormap_reference.html
data should be in matrix form like correlation or pivot table

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.head()
```

```python
tc = tips.corr()
tc
```

```python
sns.heatmap(tc,annot=True,cmap='cividis')
# uniform sequential - viridis,plasma,cividis,magma,inferno
# diverging - Spectral,coolwarm,seismic,bwr
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
fp
```

```python
# add _r to the cmap color to reverse
sns.heatmap(fp,cmap='plasma_r')
sns.heatmap(fp,cmap='plasma')
```

#### edit linewidths and linecolor

```python
sns.heatmap(fp,cmap='plasma_r',linecolor='white',linewidths=3)
```

### clustermaps put colimns and rows that are similar together

```python
sns.clustermap(fp, cmap='magma_r')
```

#### standardize the scale to 0-1 to normalize data on clustermap

```python
sns.clustermap(fp, cmap='magma_r',standard_scale=1)
```

## Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

## PairGrid 
#### gives you a lot more control than pairplot. you can choose your plots

```python
sns.PairGrid(iris)
```

```python
# Then you map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# Regression plots
lmplot allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

You can use matplotlib commands like "markers" or scatter_kws (for marker size 's') to customize

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

instead of using HUE to put sex on the same plot you can use COL to show them on separate plots. You can also use ROW to add another categorical variable, essentially creating a FacetGrid like the one we saw before

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

### change aspect (h:w) and size

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',
          aspect=0.6,size=8)
```

# Styling seaborn plots

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

### Remove all spines

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

### change size and context
use matplotlib figsize to change the size. also you can use a context to change the size to poster, notebook, talk, or paper

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster',font_scale=2)
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('notebook')
sns.countplot(x='sex',data=tips)
```
